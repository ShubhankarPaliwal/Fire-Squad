```python
# -------------------------------------------------------
# Firing Squad Synchronization
# -------------------------------------------------------

import time

# Input: number of soldiers (not including General)
soldiers = int(input("Enter number of soldiers (excluding General): "))
g_pos = int(input(f"Enter position of General (1 to {soldiers + 1}): ")) - 1

# Create line with General and Soldiers
# 'ðŸª–' = General, 'ðŸ˜' = Quiet, 'ðŸŸ¢' = Signal, 'ðŸ”´' = Reflection, 'ðŸ’¥' = Fired
line = ['ðŸ˜'] * (soldiers + 1)
line[g_pos] = 'ðŸª–'

time_step = 0
signal_phase = True
reflect_left = False
reflect_right = False

# Identify if General is at an edge
general_at_left = (g_pos == 0)
general_at_right = (g_pos == len(line) - 1)

while True:
    print(f"Time {time_step}:")
    print(' '.join(line))
    time.sleep(0.4)

    # End if all soldiers (not general) fired
    if all(s == 'ðŸ’¥' or s == 'ðŸª–' for s in line):
        break

    new_line = line.copy()

    # --------------------
    # Phase 1: Signal moves outward from General
    # --------------------
    if signal_phase:
        for i in range(len(line) - 1):
            if line[i] in ('ðŸª–', 'ðŸŸ¢') and line[i + 1] == 'ðŸ˜':
                new_line[i + 1] = 'ðŸŸ¢'
        for i in range(1, len(line)):
            if line[i] in ('ðŸª–', 'ðŸŸ¢') and line[i - 1] == 'ðŸ˜':
                new_line[i - 1] = 'ðŸŸ¢'

        # Detect when signals reach ends
        if line[0] == 'ðŸŸ¢':
            reflect_left = True
        if line[-1] == 'ðŸŸ¢':
            reflect_right = True

        # --- Start reflections depending on general position ---
        # If General is in the middle â†’ need both reflections
        # If General is at left â†’ only right reflection matters
        # If General is at right â†’ only left reflection matters
        if ((not general_at_left and not general_at_right and reflect_left and reflect_right) or    #AI
            (general_at_left and reflect_right) or
            (general_at_right and reflect_left)):

            signal_phase = False

            # Start reflection(s)
            if reflect_left:
                new_line[0] = 'ðŸ”´'
            if reflect_right:
                new_line[-1] = 'ðŸ”´'

    # --------------------
    # Phase 2: Reflections move toward the General
    # --------------------
    else:
        # Move left reflection rightwards
        for i in range(len(line) - 1):
            if line[i] == 'ðŸ”´' and line[i + 1] == 'ðŸŸ¢':
                new_line[i + 1] = 'ðŸ”´'

        # Move right reflection leftwards
        for i in range(len(line) - 1, 0, -1):
            if line[i] == 'ðŸ”´' and line[i - 1] == 'ðŸŸ¢':
                new_line[i - 1] = 'ðŸ”´'

        # When reflections reach General's neighbors, soldiers fire
        if ((g_pos == 0 and new_line[g_pos + 1] == 'ðŸ”´') or
            (g_pos == len(line) - 1 and new_line[g_pos - 1] == 'ðŸ”´')):

            for i in range(len(line)):
                if new_line[i] != 'ðŸª–':
                    new_line[i] = 'ðŸ’¥'

        elif ((g_pos > 0 and new_line[g_pos - 1] == 'ðŸ”´') and           #AI
            (g_pos < len(line) - 1 and new_line[g_pos + 1] == 'ðŸ”´')):

            for i in range(len(line)):
                if new_line[i] != 'ðŸª–':
                    new_line[i] = 'ðŸ’¥'

    line = new_line
    time_step += 1

print("\nðŸ’¥ All soldiers have fired together! ðŸ’¥")
```
